```python
#loading the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = sns.load_dataset('iris')
```

```python
data.head()
```

```python
#applying the stripplot
sns.stripplot(data=data,orient='v',palette='husl')
```

```python
#plotting a single variable
sns.stripplot(x =data['sepal_length'],data=data,orient='h')
```

```python
#here we introduce the jitter attribute which displays the distribution and can control it by passing numbers like 0.1,0.2,,,,
sns.stripplot(x='species',y='sepal_length',data=data,jitter=True)
```

```python
#here we used the new attribute dodge which split the two different plots as shown below,and marker to change the 
#type of plot marking style and the size attribute will increase its size and the alpha attribute will give you the shadow looks
sns.set(rc={'figure.figsize':(9,9)})
sns.stripplot(x='species',y='sepal_length',data=data,hue='petal_width',jitter=True,dodge=True,palette='autumn',marker='d',
              size=15,alpha=0.4)
```
