```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
%matplotlib inline
sns.set(rc={'figure.figsize':(5,5)})
```

```python
num =np.random.randn(50)
```

```python
sns.distplot(num,color='yellow')
```

```python
#if we want to add a variable to the distplot we need to define it to another variable
label_dist = pd.Series(num,name='x var')
sns.distplot(label_dist)
```

```python
#if we want to turn the distribution to vertical then type vertical=True
sns.distplot(label_dist,vertical=True)
```

```python
#if we dont want the histograms to be seen then type hist = False
sns.distplot(label_dist,hist=False)
```

```python
#instead of distplot we can see the rugplot
sns.distplot(label_dist,hist=False,rug=True,color='pink')
```
