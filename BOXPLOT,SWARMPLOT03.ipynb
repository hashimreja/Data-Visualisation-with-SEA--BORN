```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#lets plot a single variable in boxplot
sns.boxplot(x='sepal_length',data=iris,palette='husl')
```

```python
#now lets again plot all the variables here we used another argument 'orient' which is used to convert to horizontal/vertical
sns.boxplot(data=iris,palette='husl',orient='v')
```

```python
sns.boxplot(data=iris,orient='h')
```

```python
#if we want to see each and every datapoint in the boxplot then
sns.boxplot(data=iris,palette='winter')
sns.swarmplot(data=iris,color='red')
```
